```python
import subprocess
import os
os.environ["PATH"] += os.pathsep + r"C:\ffmpeg\bin"
```

```python
# 🔹 ffmpeg 경로 (환경변수 안 돼 있으면 지정)
FFMPEG = "ffmpeg"
FFPROBE = "ffprobe"
# 예: r"C:\ffmpeg\bin\ffmpeg.exe"

# 🔹 경로 설정
mp3_dir = r"C:\Users\ITSC\Desktop\파이널\original_data"
output_dir = r"C:\Users\ITSC\Desktop\파이널\head_tail_15s_cut_data"
os.makedirs(output_dir, exist_ok=True)
```

```python
CUT_SEC = 10  # 앞/뒤 제외할 초

# =========================
# mp3 처리 루프
# =========================
for file in os.listdir(mp3_dir):
    if not file.lower().endswith(".mp3"):
        continue

    input_path = os.path.join(mp3_dir, file)
    output_path = os.path.join(output_dir, file)

    print(f"▶ 처리중: {file}")

    try:
        # 1️⃣ 전체 길이 구하기
        cmd_duration = [
            "ffprobe", "-i", input_path,
            "-show_entries", "format=duration",
            "-v", "quiet",
            "-of", "csv=p=0"
        ]
        duration = float(subprocess.check_output(cmd_duration).decode().strip())

        # 길이가 너무 짧은 파일 보호
        if duration <= CUT_SEC * 2:
            print(f"⚠️ 스킵 (길이 {round(duration,2)}초)")
            continue

        # 2️⃣ 앞/뒤 15초 제외하고 가운데만 컷
        cmd_cut = [
            "ffmpeg", "-y",
            "-i", input_path,
            "-ss", str(CUT_SEC),
            "-to", str(duration - CUT_SEC),
            "-c", "copy",
            output_path
        ]

        subprocess.run(
            cmd_cut,
            stdout=subprocess.DEVNULL,
            stderr=subprocess.DEVNULL
        )

    except Exception as e:
        print(f"❌ 실패: {file}")
        print(e)

print("\n✅ 앞/뒤 15초 제외 mp3 생성 완료")
```
